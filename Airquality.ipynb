```python
# Install required libraries (if not already)
!pip install seaborn

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file (make sure you've uploaded it in the left panel)
df = pd.read_csv('air_quality_dataset (1).csv')  # Make sure this name matches your uploaded file

# Display first 5 rows
print("Preview of the dataset:")
display(df.head())

# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Line Plot - AQI over Date
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['AQI'], marker='o', color='green')
plt.title('Air Quality Index (AQI) Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter Plot - Temperature vs AQI
plt.figure(figsize=(6,4))
plt.scatter(df['Temperature'], df['AQI'], color='blue')
plt.title('Temperature vs AQI')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('AQI')
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8,5))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap Between Features')
plt.tight_layout()
plt.show()
```
